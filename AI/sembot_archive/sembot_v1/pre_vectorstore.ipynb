```python
import json
from pathlib import Path
from pprint import pprint
```

```python
file_path = "../kisa_json/급여규정.json"
```

```python
file_path.split('/')[-1].split('.')[0]
```

```python

json_data = json.loads(Path(file_path).read_text())

# pprint(data)
```

```python
pages = json_data["page"]
```

```python
new_datas = []
```

```python
article_seq = 0
page_title = ""

for index, page in enumerate(pages, 1):
    page_title = f"제{index}장({page['content']})"
    
    new_data = {}


    for article in page["article"]:
        article_seq += 1
        content = ""

        meta = f"{page_title} 제{article_seq}조({article['title']})"

        # article의 content가 있으면 추가
        if article.get('content'):
            content += article['content'] + " "

        # 항이 있으면 추가
        if article.get('paragraph'):
            for para_index, paragraph in enumerate(article['paragraph'], 1):
                content += f"제{para_index}항 {paragraph['content']}\n"

                # 호가 있으면 추가
                if paragraph.get('sub-paragraph'):
                    for subPara_index, sub_paragraph in enumerate(paragraph['sub-paragraph'], 1):
                        content += f"제{subPara_index}호 {sub_paragraph['content']}\n"


        new_data['meta'] = meta
        new_data['content'] = content

        pprint(new_data)



    print("======================================")

new_datas.append(new_data)
```
